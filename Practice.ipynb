```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
# Scrape Across Pages
for x in range(1, 6): #use range(1, 6) in our for loop to visit the first five pages of the website
   html = browser.html
   quote_soup = soup(html, 'html.parser') # 1. Create a BeautifulSoup object,
   quotes = quote_soup.find_all('span', class_='text') # 2. Find all the quotes on the page 
   for quote in quotes: 
      print('page:', x, '----------')
      print(quote.text) #3. Print each quote from the page
   browser.links.find_by_partial_text('Next').click() # 4. Click the "Next" button at the bottom of the page
```

```python
# SKILL DRILL http://books.toscrape.com/ Stretch your scraping skills by visiting Books to Scrape (Links to an external site.) and scraping the book URL list on the first page
# Visit the Quotes to Scrape site
url = 'https://newjersey.craigslist.org/search/sss?sort=rel&query=guitar'
browser.visit(url)
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = soup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('li', class_="result-row")
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="result-title").text
        # Identify and return price of listing
        price = result.a.span.text
        # Identify and return link to listing
        link = result.a['href']

        # Print results only if title, price, and link are available
        if (title and price and link):
            print('-------------')
            print(title)
            print(price)
            print(link)
    except AttributeError as e:
        print(e)
```
